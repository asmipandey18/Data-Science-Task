```python
#Hypothesis Formulation and Statistical Testing
#Research Question
#Is there a significant relationship between a house's living area and its selling price

#Hypotheses
#Null Hypothesis (H₀):
#There is no significant relationship between living area (sqft_living) and house price (price).

#Alternative Hypothesis (H₁):
#There is a significant relationship between living area (sqft_living) and house price (price).

#Statistical Test
#Pearson Correlation Test is used because both variables are numerical.
corr, p_value = stats.pearsonr(df['sqft_living'], df['price'])

print("Correlation Coefficient:", corr)
print("P-value:", p_value)
```

```python
#Decision Rule
#Interpertation
#p-value < 0.05 = significance relationship exists.
#p-value > 0.05 = No significance relationship.
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```
